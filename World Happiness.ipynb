```python
from sklearn.cluster import KMeans

import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
sns.distplot(df['GDP per capita'])
```

```python
sns.distplot(df['Score'])
```

```python
# dataframe that will be used in algorithm
X = df[['Score', 'GDP per capita', 'Healthy life expectancy']]
X.head()
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X)
```

```python
len(df)
```

```python
kmeans.cluster_centers_
```

```python
df['cluster'] = kmeans.labels_
df.head()
```

```python
sns.pairplot(data=df, hue='cluster')
```

```python
# compare "Income" and "Score"
sns.scatterplot(data=df, x='GDP per capita', y='Healthy life expectancy', hue='cluster')

# scatterplot of centroid coordinates for income and spending score
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], c='red', marker='x')

plt.title("GDP per capita and Healthy Life Expectancy Groups")
#put the legend on the right
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
country = [32, 75, 68]

country_group = kmeans.predict([country])

print(f"Cluster assignment: {country_group}")
```

```python
country = [0.50, 0.75, 4]

country_group = kmeans.predict([country])

print(f"Cluster assignment: {country_group}")
```

```python
from sklearn.naive_bayes import GaussianNB   #import Gaussian Bayes modeling function
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df.count()
```

```python
df.dtypes
```

```python
df['Happy'] = np.where(df['Score']>= 6, 1, 0)
```

```python
#dataframe with predicting features
X = df.drop('Happy', axis=1)

#column of predictive target values
y = df['Happy']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=109)
```

```python
gnb = GaussianNB()
```

```python
df.drop('Country or region', axis=1, inplace=True)
```

```python
gnb.fit(X_train, y_train)
```

```python
gnb.score(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
cm = pd.DataFrame(
    confusion_matrix(y_test, y_pred),
    columns=['Predicted Failed', 'Predicted Passed'],
    index=['True Failed', 'True Passed']
)

cm
```

```python
gnb.score(X_test, y_test)
```
